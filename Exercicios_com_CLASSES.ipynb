## Estrutura de repetição



A seguir vamos tratar um erro no Python. TAmbém podemos entender o seguinte trecho de código como um código, na qual queremos que o usuário digite um valor válido. Vamos lá:

```python
while True:
  
  try:
    numero = int(input("Digite um valor de 0 a 10: "))
  except ValueError:
    print("Deve ser fornecido um valor inteiro")
  else:
    if 0<= numero <= 10:
      print(f"Número informado é: {numero}")
      break
    else:
      print("O número deve estar entre 0 e 10")  


```

Agora sim, vamos criar nosso método

```python
# Vamos criar nosso método dentro da nossa classe

class CirculoPerfeito():
  def __init__(self):
    self.cor = "Azul"
    self.circunferencia = 4
    self.material = "papel"


  #def mostra_cor(self):
    #return id(self)
  def mostra_cor(self):
    return self.cor  

  def trocar_cor(self, cor):
    self.cor = cor

```

```python
circulo_primeiro = CirculoPerfeito()
circulo_segundo = CirculoPerfeito()

#circulo_segundo.cor = "Amarelo" 
circulo_segundo.trocar_cor("Amarelo")


print(id(circulo_primeiro), circulo_primeiro.mostra_cor())
print(id(circulo_segundo), circulo_segundo.mostra_cor())


print(circulo_primeiro.cor, circulo_segundo.cor)
```

```python
# Posso alterar a cor do atributo

circulo_segundo.cor = "Verde"
print(circulo_primeiro.cor, circulo_segundo.cor)
```
